```python
#класс  бор с методами добавления, поиска и удаления
def create_node():
    return {'children': {}, 'is_end': False}

def add(root, word):
    current = root
    for char in word:
        if char not in current['children']:
            current['children'][char] = create_node()
        current = current['children'][char]
    current['is_end'] = True

def search(root, word):
    current = root
    for char in word:
        if char not in current['children']:
            return False
        current = current['children'][char]
    return current['is_end']

def delete(root, word):
    def _delete(node, word, depth):
        if depth == len(word):
            if not node['is_end']:
                return False  # Слово не найдено
            node['is_end'] = False
            # Если нет дочерних узлов, можно удалить этот узел
            return len(node['children']) == 0
        char = word[depth]
        if char not in node['children']:
            return False  # Слово не найдено
        should_delete_child = _delete(node['children'][char], word, depth + 1)
        if should_delete_child:
            del node['children'][char]
            # Удаляем текущий узел, если он не является концом слова и не имеет детей
            return not node['is_end'] and len(node['children']) == 0
        return False

    return _delete(root, word, 0)
```

```python
#Посчитать количество вхождений строки и всех ее циклических сдвигов в текст
def count_occurrences(text, pattern):

    n = len(pattern)
    if n == 0:
        return 0

    # Создаём множество всех циклических сдвигов pattern
    shifts = set()
    doubled = pattern + pattern  # для удобного получения сдвигов
    for i in range(n):
        shifts.add(doubled[i:i+n])

    total = 0
    # Для каждого сдвига ищем все вхождения в text
    for s in shifts:
        start = 0
        while True:
            pos = text.find(s, start)
            if pos == -1:
                break
            total += 1
            start = pos + 1  # сдвигаемся на 1 символ, чтобы учитывать пересекающиеся вхождения

    return total

```

```python
def restore_from_z(z):
    """
    Восстанавливает строку по Z-функции.
    Возвращает лексикографически минимальную строку из символов 'a' и 'b'.
    Выбрасывает ValueError, если Z-функция некорректна.
    """
    n = len(z)
    if n == 0:
        return ""

    # Инициализируем строку символами 'a'
    s = ['a'] * n

    # Проверка корректности Z[0]
    if z[0] != 0:
        raise ValueError("Z[0] должен быть равен 0")

    for i in range(1, n):
        if z[i] < 0 or z[i] > n - i:
            raise ValueError(f"Недопустимое значение Z[{i}] = {z[i]}")

        if z[i] > 0:
            # Копируем символы из начала строки в позицию i
            for j in range(z[i]):
                pos = i + j
                if s[pos] == 'a':
                    s[pos] = s[j]
                elif s[pos] != s[j]:
                    raise ValueError(f"Конфликт символов на позиции {pos}")
        else:
            # z[i] == 0, символ s[i] должен отличаться от s[0]
            if s[i] == 'a':
                s[i] = 'b'
            elif s[i] == s[0]:
                raise ValueError(f"Символ на позиции {i} должен отличаться от начала строки")

    return "".join(s)
```
